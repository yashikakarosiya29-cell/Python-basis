2) What is an interpreter in Python?
-  In Python, an interpreter is the program that reads and executes python code. an interpreter processes code line by line, translating it into intermediate instructions and executing them immediately.


3) What are pre-defined keywords in Python ?
-  In Python, predefined keywords are the reserved words that have special meaning in the language. These words are "built in" and you cannot use them as identifiers ,because python uses them to define its own syntax and structure.
Examples of Keywords in Python:
  1) Control flow: if, else, elif ,while, for, break
  2) Data handling: true ,false, none, and, or,not
  3) Functions & classes: def ,return ,class
  4) imports & modules: import, from, as, global

5) What is mutability in Python ?
-  Mutability refers to whether the value of an object can be changed after it is created.


6) Why are lists mutable, but tuples are immutable?
-  Lists - Lists are meant to be flexible, dynamic containers. They are designed so you can add ,remove ,or change elements after the list is created.

   Tuples - Tuples are meant to be fixed collections of items. Once created, their contents cannot be changed.

10) What is the difference between implicit and explicit type casting?
- 1) Implicit type casting (type promotion)

  - Done automatically by Python.
   - Happens when you mix two types and Python converts the smaller type into the larger type to avoid data loss.
  2) Explicit type casting
    - Done manually using built in functions.
    - we dedide the target data type.

13. What is the difference between for and while loops ?
-  **For loop** -
  . used when you know how  many times you want to loop, or when looping through a sequence (like a list, string, or range)
  - the loop runs for each item in a collection.

   While loop-
   - used when you don't know exactly how many times to loop in advahce.
   - it keeps running as long as a condition is True
   - you must make sure the condition eventually becomes False, or you'll get an infinite loop.





14) Describe a scenario where a while loop is more suitable than a for loop.
-   A while loop is more suitable when we do not know in advance how many times the loop will run.
 for example- if we keep asking the user to enter a password until they enter a correct one ,we cannot decide the number of attempts before the loop starts. In such cases, a while loop is better than a for loop.
